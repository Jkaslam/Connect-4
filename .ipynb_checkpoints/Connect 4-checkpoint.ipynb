```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
import sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

data = pd.read_csv("connect-4data.csv")

# Make the data numerical by turning b -> 1, x-> 2, o -> 3
data.replace(to_replace =['b', 'x', 'o', 'win', 'loss', 'draw'], value=[1, 2, 3, 1, 2, 3], inplace=True)

features = data.iloc[:, :-1]

# Hot encode the labels, i.e. convert the integers representing classes to binary
hot_encoder = OneHotEncoder();
labels = hot_encoder.fit_transform(data.iloc[:, [-1]]).toarray()

for k in range(9):
    model_accuracies = list()
    for j in range(100):
        # Split data into training data and testing data
        train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = .9 - (k * .1))

        # Configure the neural network
        model = Sequential()
        model.add(Dense(40, input_dim=42, activation='relu'))
        model.add(Dense(40, activation='relu'))
        model.add(Dense(3, activation='softmax'))
        model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

        # Train the neural network on the input data
        history = model.fit(train_features.to_numpy(), train_labels, epochs=100, batch_size = 64, verbose=0)

        # Figures out the accuracy of the current model
        pred_softmax = model.predict(test_features)
        pred_labels = list()
        for i in range(len(pred_softmax)):
            pred_labels.append(np.argmax(pred_softmax[i]))
        unencoded_test_labels = list()
        for i in range(len(test_labels)):
            unencoded_test_labels.append(np.argmax(test_labels[i]))
    
        test_accuracy = accuracy_score(pred_labels, unencoded_test_labels)
        #print(test_accuracy)
        model_accuracies.append(test_accuracy)

    plt.scatter(range(100), model_accuracies)
    plt.figure()
plt.show()
```

```python
from sklearn import tree
from sklearn import metrics

x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size = .9)

classifier = tree.DecisionTreeClassifier(max_depth=15)
classifier.fit(x_train, y_train)

pred = classifier.predict(x_test)
print("Accuracy of decision tree is:", metrics.accuracy_score(y_test, pred))

```

```python
import copy

def count_threes(curr_player, board):
    num_of_3 = 0
    #Loop through columns
    for i in range(7):
        #Loop through rows
        for j in range(6):
            current_cell = board[i][j]
            if current_cell == 'b' or current_cell != curr_player:
                continue
            #Check right
            if i < 5 and current_cell == board[i+1][j] == board[i+2][j]:
                num_of_3 += 1
            #Check left
            if i > 1 and current_cell == board[i-1][j] == board[i-2][j]:
                num_of_3 += 1
            #Check up 
            if j < 4 and current_cell == board[i][j+1] == board[i][j+2]:
                num_of_3 += 1
            #Check down
            if j > 1 and current_cell == board[i][j-1] == board[i][j-2]:
                num_of_3 += 1
            #Check diagonal up and right
            if i < 5 and j < 4 and current_cell == board[i + 1][j+1] == board[i+2][j+2]:
                num_of_3 += 1
            #Check diagonal up and left
            if i > 1 and j < 4 and current_cell == board[i-1][j+1] == board[i-2][j+2]:
                num_of_3 += 1
            #Check diagonal down and right
            if j > 1 and i < 5 and current_cell == board[i+1][j-1] == board[i+2][j-2]:
                num_of_3 += 1
            #Check diagonal down and left
            if j > 1 and i > 1 and current_cell == board[i-1][j-1] == board[i-2][j-2]:
                num_of_3 += 1

    # Since we count from both directions we will double count runs of 3
    num_of_3 /= 2
            
    return(num_of_3)
        
def x_potential_threes(board):
    x_curr_threes = count_threes('x', board)
    print(x_curr_threes)
    x_pot_threes = 0
    
    #Loop through columns
    for i in range(7):
        #Loop through rows
        for j in range(6):
            if board[i][j] == 'b':
                board2 = copy.deepcopy(board)
                board2[i][j] = 'x'
                print(board2)
                x_pot_threes += count_threes('x', board2) - x_curr_threes
    return x_pot_threes
            
input_board = [['x', 'x', 'b', 'b', 'b', 'b'], ['x', 'x', 'b', 'b', 'b', 'b'], ['b', 'b', 'b', 'b', 'b', 'b'], ['b', 'b', 'b', 'b', 'b', 'b'], ['b', 'b', 'b', 'b', 'b', 'b'], ['b', 'b', 'b', 'b', 'b', 'b'], ['b', 'b', 'b', 'b', 'b', 'b']]
x_potential_threes(input_board)
```
